```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('128','256','300','512')
y_pos = np.arange(len(objects))
performance = [0.08519048653,0.0917349344,0.0997605046,0.07922477858]
 
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.xlabel('Word Embedding Dimension')
plt.ylabel('Spearman Correlation Score')
plt.title('Performance of Word Embedding vs Dimension')
plt.savefig('Word_Emb_Dim.png')
plt.show()
```
